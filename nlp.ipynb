```python
#unique emotion
unique_emotions = df['emotion'].unique()
emotion_numbers = {}
i = 0
for emo in unique_emotions:
  emotion_numbers[emo] = i
  i += 1

df['emotion'] = df['emotion'].map(emotion_numbers)
```

```python
df['text'] =df['text'].apply(lambda x : x.lower())
```

```python
import string
def remove_punc(text):
  for punc in string.punctuation:
    text = text.replace(punc,'')
  return text
```

```python
def remove_numbers(txt):
  new =""
  for i in txt:
    if not i.isdigit():
      new += i
  return new

df['text'] = df['text'].apply(remove_numbers)
```

```python
def remove_emojis(txt):
  new =""
  for i in txt:
    if i.isascii(): # Corrected: Keep ASCII characters, thus removing non-ASCII (emojis)
      new += i
  return new

df['text'] = df['text'].apply(remove_emojis)
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
stop_words = set(stopwords.words('english'))
len(stop_words)
```

```python
stop_words
```

```python
df.loc[1]['text']
```

```python
def remove(txt):
  words = txt.split()
  new_words = []
  for i in words:
    if i not in stop_words:
      new_words.append(i)
  return ' '.join(new_words)
```

```python
df['text'] = df['text'].apply(remove)
```

```python
from sklearn.model_selection import train_test_split

# Filter out rows where 'text' is an empty string after preprocessing
df_filtered = df[df['text'].str.strip() != '']

X = df_filtered['text']
y = df_filtered['emotion']

X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=42)
```

```python
X_test
```

```python
df.shape
```

```python
from sklearn.feature_extraction.text import TfidfTransformer ,CountVectorizer
```

```python
bow_vectorizer = CountVectorizer()
```

```python
X_train_bow = bow_vectorizer.fit_transform(X_train)
X_test_bow = bow_vectorizer.transform(X_test) # Corrected: Use transform for the test set
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
nb_model = MultinomialNB()
nb_model.fit(X_train_bow, y_train)
```

```python
pred_nb = nb_model.predict(X_test_bow)
print(accuracy_score(y_test, pred_nb))
```

```python
y_test
```

```python
tfidf_vectorizer = TfidfTransformer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train_bow)
X_test_tfidf = tfidf_vectorizer.transform(X_test_bow)

nb_model_tfidf = MultinomialNB()
nb_model_tfidf.fit(X_train_tfidf, y_train)
pred_nb_tfidf = nb_model_tfidf.predict(X_test_tfidf)
```

```python
print(accuracy_score(y_test, pred_nb_tfidf))
```

```python
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression(max_iter=1000)
logistic_model.fit(X_train_tfidf, y_train)
pred_logistic = logistic_model.predict(X_test_tfidf)
```

```python
print(accuracy_score(y_test, pred_logistic))
```
